# 1. Data Preparation:

```python
import pandas as pd
from PIL import Image
import numpy as np
```

```python
# Features and target variable
X = data[["ID",'Thinckness', 'Weight', 'Surface', 'Color', 'Transparence']]
y = data['Material']
```

```python
# Image loading function
def load_image(image_id, base_path='Camera2/front'):
    # Replace with the path to your images directory
    image_path = f"{base_path}/{image_id}.jpeg"
    try:
        with Image.open(image_path) as img:
            img = img.resize((128, 128))  # Resize image
            return np.array(img)
    except FileNotFoundError:
        # Return NaN or a placeholder image (e.g., all zeros)
        print(f"Image file not found: {image_path}")
        return np.full((128, 128, 3), np.nan)  # Return a placeholder image with NaN values
# Load images
images = np.array([load_image(image_id) for image_id in data['ID']])
```

# 2. Data Preprocessing:

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.models import Model
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

### Feature Extraction:

```python
# Load a pre-trained CNN model for feature extraction
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
model = Model(inputs=base_model.input, outputs=base_model.output)
```

```python
def extract_features_from_images(images):
    features = []
    for img in images:
        if np.isnan(img).any():  # Check if the image contains NaN values
            features.append(np.zeros((4 * 4 * 512,)))  # Return a zero-filled vector as placeholder
        else:
            img = preprocess_input(img)
            img = np.expand_dims(img, axis=0)
            feature = model.predict(img)
            features.append(feature.flatten())
    return np.array(features)

# Extract image features
image_features = extract_features_from_images(images)
```

```python
if image_features.ndim == 3:
    # Flatten the image features to 2D: [n_samples, height * width * channels]
    image_features = image_features.reshape(image_features.shape[0], -1)
```

# 3. Model Training:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import FeatureUnion
from sklearn.model_selection import train_test_split
```

```python
# Define numerical and categorical features
numeric_features = ['Thinckness', 'Weight', 'Surface']
categorical_features = ['Color', 'Transparence']
target = 'Material'
```

### Pipeline:

```python
# Preprocessing for numerical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Handle missing values
    ('scaler', StandardScaler())  # Normalize numerical data
])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Handle missing values
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encode categorical data
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

numeric_categorical_features = preprocessor.fit_transform(data[numeric_features + categorical_features])
```

```python
# Ensure both feature arrays have the same number of samples and are 2D
assert numeric_categorical_features.shape[0] == image_features.shape[0], "Mismatch in number of samples"
assert numeric_categorical_features.ndim == 2, "Numeric/Categorical features should be 2D"
assert image_features.ndim == 2, "Image features should be 2D"
```

```python
from scipy.sparse import csr_matrix
# Convert sparse matrix to dense array
if isinstance(numeric_categorical_features, csr_matrix):
    numeric_categorical_features = numeric_categorical_features.toarray()

# Debugging shapes
print("Numeric/Categorical Features Shape:", numeric_categorical_features.shape)
print("Image Features Shape:", image_features.shape)
```

```python
# Combine numerical, categorical, and image features
combined_features = np.hstack([
    numeric_categorical_features,
    image_features
])
```

# 4. Weight Fixing:

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
# Create a DataFrame with the target variable and features
df_combined = pd.DataFrame(combined_features)
df_combined[target] = y.values  # Add target variable to the DataFrame

# Balance each class by resampling
df_balanced = pd.DataFrame()

for material in df_combined[target].unique():
    class_df = df_combined[df_combined[target] == material]
    
    # Find the maximum number of samples
    max_samples = df_combined[target].value_counts().max()
    
    # Resample with replacement to achieve the maximum number of samples
    resampled_class_df = resample(class_df,
                                  replace=True,
                                  n_samples=max_samples,
                                  random_state=1)
    
    # Append resampled data to the balanced DataFrame
    df_balanced = pd.concat([df_balanced, resampled_class_df], axis=0)

# Shuffle the balanced DataFrame
df_balanced = df_balanced.sample(frac=1, random_state=1).reset_index(drop=True)

# Separate features and target after balancing
X_balanced = df_balanced.drop(columns=[target])
y_balanced = df_balanced[target]
```

```python
# Split balanced data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)
print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```
