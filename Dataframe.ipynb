```python
import pandas as pd

# Path to the Excel file
file_path = 'use.xlsx'
date_format = '%Y-%m-%d %H:%M:%S'
# Read the specific sheet named 'Daten' into a dataframe
# Use the second row (index 1) as the header and skip the first six rows
df = pd.read_excel(file_path, sheet_name='Daten', header=1, parse_dates=True, date_parser=lambda x: pd.to_datetime(x, format=date_format))

# Print the dataframe
df
```

```python
# Drop the first 5 rows of the dataframe
df = df.iloc[5:].reset_index(drop=True)
```

```python
df['Time'] = df['Time'].astype(str)
```

```python
# Save the dataframe as clean.xlsx
df.to_excel('clean.xlsx', index=False)
```
